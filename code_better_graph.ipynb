We used the *Game results for Top 500 Players from 2017 - 2019* Dataset, and are using the dataset to classify what Tournament Level a French game is being played at, depending on
the number of aces, as well as the number of successful first serves.

```r
# Loading the Tidyverse Library
library(tidyverse)
```

```r
# Reading the Data from Relative Storage
data <- read_csv("data/atp2017-2019.csv")
data
```

```r
# Cleaning the Data

options(repr.plot.width = 12, repr.plot.height = 12)

clean_data <- select(data, tourney_level, winner_ioc, surface, w_ace, w_1stIn, l_ace, l_1stIn) %>%
    filter(surface == "Hard"| surface == "Grass"| surface == "Clay") %>%
    filter(surface == "Hard") %>%
    filter(winner_ioc == "FRA") %>%
    filter(!is.na(w_ace))  %>%
    mutate(tourney_level = as_factor(tourney_level)) %>%
    mutate(ace = w_ace + l_ace) %>%
    mutate(successful_serves = w_1stIn + l_1stIn)

clean_data

# Graphing the Data


plot_data <- ggplot(clean_data, aes(x = ace, y = successful_serves, color = tourney_level)) + 
    geom_point(aes(color = tourney_level)) +
    ggtitle("Classification of Tournament Level in France based on Aces and Successful First Serves") +
    labs(x = "Aces (Winner & Loser)", y = "Number of successful first serves (Winner & Loser)", color = "Tournament Level") + 
    theme(text = element_text(size = 15))


plot_data
```
